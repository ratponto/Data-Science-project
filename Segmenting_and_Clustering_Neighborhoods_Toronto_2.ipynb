```python
import numpy as np
import pandas as pd
```

**Import neighborhoods data from first task**

```python
toronto_data = "Toronto_neighborhoods.csv"
TorontoPostalCodes = pd.read_csv(toronto_data).set_index('Postal code')
TorontoPostalCodes.head()
```

**Use Google Maps Geocoding API**

*Leveraging the Google Maps Geocoding API to get the latitude and the longitude coordinates of each neighborhood in order to utilize the Foursquare location data.*
- extract csv with Tornto geographical coordinates to dataframe

```python
toronto_geocsv = 'https://cocl.us/Geospatial_data'
!wget -q -O 'toronto_m.geospatial_data.csv' 'https://cocl.us/Geospatial_data'
geocsv_data = pd.read_csv(toronto_geocsv).set_index("Postal Code")
geocsv_data.head()
```

**Join tables**

*Combining the two dataframes for the final result*

```python
toronto_neighborhoods = TorontoPostalCodes.join(geocsv_data)
```

```python
toronto_neighborhoods
```

```python
toronto_neighborhoods.to_csv('Toronto_neighborhoods_v2.csv',index=True)
```
